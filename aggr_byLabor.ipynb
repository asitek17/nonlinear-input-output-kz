Агрегация в большие области

```python
def aggregation_big_sph(A, Y, aggr_list):
    s = len(aggr_list)
    B = np.zeros([s, s])
    Y_new = np.zeros(s)
    for (k, num_i) in enumerate(aggr_list):
        for (l, num_j) in enumerate(aggr_list):
            x_j = Y.iloc[num_j]
            tmp = A.iloc[num_i, :]
            a_ij = tmp.iloc[:, num_j]
            B[k, l] = np.sum(a_ij @ x_j) / np.sum(x_j)
        x_j = Y.iloc[num_i]
        Y_new[k] = np.sum(x_j)
    return B, Y_new
```
