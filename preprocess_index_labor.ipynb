```python
import numpy as np
import pandas as pd
```

```python
pre = './data/3q_info/'
file_name = pre + 'index_labor_pre.ods'


series_labor_pre =  pd.read_excel(file_name, index_col=0, engine='odf')
series_labor_pre_norm = series_labor_pre.div(series_labor_pre.iloc[:, 0], axis=0).fillna(0)
series_labor = series_labor_pre_norm.apply(lambda row: row.cumprod(), axis=1)
```

```python
series_labor
```

```python
year = 2017
ind = year - 2017
series_labor_my = series_labor.div(series_labor.iloc[:, ind], axis=0).fillna(0)
```

```python
with pd.ExcelWriter(pre + 'series_labor' + str(year) + '.xlsx') as writer:
    series_labor_my.to_excel(writer)

print("Данные успешно записаны в файл")
```

```python
pre = './data/3q_info/'
file_name = pre + 'index_import.ods'

import_val =  pd.read_excel(file_name, index_col=0, engine='odf')
row_numbers = np.arange(len(import_val))
mask = row_numbers % 3 == 0
import_val_new = import_val[mask]
```

```python
import_val_new2 = import_val_new.copy()
for col in import_val_new2.columns:
    import_val_new2[col] = import_val_new2[col].astype(str).str.replace(',', '.')
    
    # Преобразуем в числовой формат
    import_val_new2[col] = pd.to_numeric(import_val_new2[col])

# Проверяем типы данных
print(import_val_new2.dtypes)
```

```python
import_ = import_val_new2.div(import_val_new2.iloc[:, 0], axis=0).fillna(0)
import_2 = import_.apply(lambda row: row.cumprod(), axis=1)
```

```python

with pd.ExcelWriter(pre + 'test_import.xlsx') as writer:
    import_2.to_excel(writer, index=True)

print("Данные успешно записаны в файл")
```
